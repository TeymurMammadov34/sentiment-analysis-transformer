# ✅ Gerekli Kütüphanelerin Yüklenmesi

```python
# Sayısal işlemler için NumPy kütüphanesi içe aktarılıyor.
import numpy as np

# TensorFlow ve Keras modülleri içe aktarılıyor.
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Görselleştirme için Matplotlib kullanılacak.
import matplotlib.pyplot as plt

# Gereksiz uyarı mesajlarını bastırmak için warnings modülü kullanılıyor.
import warnings
warnings.filterwarnings("ignore")
```

# ✅ IMDb Verisinin Yüklenmesi ve İşlenmesi

```python
# En sık kullanılan 10.000 kelime ile sınırlama yapılır.
max_features = 10000

# Her yorum en fazla 100 kelime uzunluğunda olacak şekilde ayarlanır.
maxlen = 100

# IMDb veri seti yüklenir. Sadece en sık geçen 10.000 kelime dahil edilir.
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Eğitim ve test verileri aynı uzunlukta olacak şekilde pad edilir (kısa olanlar <PAD> ile doldurulur).
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# IMDb verisinde kullanılan kelimelerin indeks karşılıkları alınır.
word_index = imdb.get_word_index()

# İndeks-kelime sözlüğü ters çevrilir. Keras’ın özel token’ları nedeniyle +3 yapılır.
reverse_word_index = {index + 3: word for word, index in word_index.items()}

# Özel semboller atanır:
reverse_word_index[0] = "<PAD>"
reverse_word_index[1] = "<START>"
reverse_word_index[2] = "<UNK>"
reverse_word_index[3] = "<UNUSED>"
```

# ✅ Yorumların Decode Edilmesi ve Gösterilmesi

```python
# Sayısal verileri tekrar metin haline çevirir.
def decode_review(encoded_review):
    return " ".join([reverse_word_index.get(i, "?") for i in encoded_review])

# Eğitim setinden rastgele 3 yorum seçilir.
random_indices = np.random.choice(len(x_train), size=3, replace=False)

# Seçilen yorumlar ekrana yazdırılır.
for i in random_indices:
    print(f"Yorum: {decode_review(x_train[i])}")
    print(f"Etiket: {y_train[i]}")
    print()
```

# 🧱 TransformerBlock Sınıfı

```python
# Transformer bloğunun tanımı
class TransformerBlock(layers.Layer):

    def __init__(self, embed_size, heads, dropout_rate=0.3):
        super(TransformerBlock, self).__init__()

        # Çok başlı dikkat mekanizması (self-attention)
        self.attention = layers.MultiHeadAttention(num_heads=heads, key_dim=embed_size)

        # Normalizasyon katmanları
        self.norm1 = layers.LayerNormalization(epsilon=1e-6)
        self.norm2 = layers.LayerNormalization(epsilon=1e-6)

        # Feed forward ağı (2 tam bağlantılı katman)
        self.feed_forward = models.Sequential([
            layers.Dense(embed_size * 4, activation="relu"),  # Gizli katman
            layers.Dense(embed_size)  # Giriş boyutuna geri dönüş
        ])

        # Overfitting'i azaltmak için dropout
        self.dropout1 = layers.Dropout(dropout_rate)
        self.dropout2 = layers.Dropout(dropout_rate)

    def call(self, x, training):
        # Self-attention uygulanır
        attention = self.attention(x, x, training=training)

        # İlk normalizasyon ve residual connection
        x = self.norm1(x + self.dropout1(attention, training=training))

        # Feed forward + dropout + ikinci normalizasyon
        feed_forward = self.feed_forward(x)
        return self.norm2(x + self.dropout2(feed_forward, training=training))
```

# 📦 TransformerModel

```python
# Kendi Transformer tabanlı sınıflandırma modelimizi tanımlıyoruz
class TransformerModel(models.Model):

    def __init__(self, num_layers, embed_size, heads, input_dim, output_dim, dropout_rate = 0.2):
        super(TransformerModel, self).__init__()

        # 🔤 Embedding katmanı: kelimeleri sayılardan vektörlere dönüştürür
        self.embedding = layers.Embedding(input_dim=input_dim, output_dim=embed_size)

        # 🧱 Birden fazla Transformer bloğu oluşturuluyor (örneğin 4 katman)
        self.transformer_blocks = [
            TransformerBlock(embed_size, heads, dropout_rate)
            for _ in range(num_layers)
        ]

        # 📏 Zaman boyutunu ortalayıp sabit vektör çıkarmak için global average pooling
        self.global_avg_pooling = layers.GlobalAveragePooling1D()

        # 💧 Overfitting'i önlemek için dropout uygulaması
        self.dropout = layers.Dropout(dropout_rate)

        # 🔚 Sonuç katmanı: 1 çıktı üretir (sigmoid ile 0-1 arası değer — binary classification için)
        self.fc = layers.Dense(output_dim, activation="sigmoid")

    def call(self, x, training):
        # 🔤 Giriş verisini embedding vektörlerine çevir
        x = self.embedding(x)

        # 🔁 Her transformer bloğunu sırayla uygula
        for transformer in self.transformer_blocks:
            x = transformer(x, training=training)

        # 📏 Zamansal boyutu ortalama alarak sabit uzunlukta tek bir vektöre indir
        x = self.global_avg_pooling(x)

        # 💧 Eğitim sırasında dropout uygula (overfitting'e karşı)
        x = self.dropout(x, training=training)

        # 🔚 Son tahmin katmanı (sigmoid çıktısı)
        return self.fc(x)
```

# ⚙️ Transformer Modeli Oluşturma ve Derleme

```python
# 🔢 Transformer mimarisi için temel hiperparametreleri tanımlıyoruz
num_layers = 4            # Kaç adet Transformer bloğu kullanılacak
embed_size = 64           # Her kelimenin embedding vektör boyutu
num_heads = 4             # Multi-head attention'da kaç 'baş' kullanılacak
input_dim = max_features  # Modelin göreceği maksimum kelime sayısı (sözlük boyutu)
output_dim = 1            # Binary sınıflandırma için tek bir çıkış (0 veya 1)
dropout_rate = 0.3        # Dropout oranı (%30)

# 🧠 Transformer modelini oluşturuyoruz
model = TransformerModel(num_layers, embed_size, num_heads, input_dim, output_dim, dropout_rate)

# 🏗️ Modelin input şekli belirtilerek yapısı oluşturuluyor (embedding + attention için gerekli)
model.build(input_shape=(None, maxlen))  # (None, maxlen): batch_size belirsiz, sabit dizi uzunluğu

# 🛠️ Modeli derliyoruz:
# - Kayıp fonksiyonu: binary_crossentropy (çünkü çıktı 0 veya 1)
# - Optimizasyon: Adam
# - Performans metriği: Accuracy (doğruluk)
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

# 🏋️‍♂️ Transformer Modelini Eğitme

```python
# 🏁 Modeli eğitiyoruz:
# - x_train: eğitim verileri
# - y_train: eğitim etiketleri
# - epochs=5: model veriyi 5 kez görecek
# - batch_size=256: her seferinde 256 örnekle ağırlıklar güncellenecek
# - validation_data: eğitim sırasında doğrulama (test) verileriyle kontrol edilecek
history = model.fit(
    x_train, y_train,              # Eğitim verisi ve etiketleri
    epochs=5,                     # 5 defa tüm veri üzerinden geç
    batch_size=256,                # 256 veri ile bir batch oluştur
    validation_data=(x_test, y_test)  # Doğrulama için test verisi kullan
)
```

# 📈 Eğitim Sürecinin Görselleştirilmesi 

```python
# history nesnesinin içindeki gerçek verilere erişiyoruz (sadece dictionary kısmı lazım)
history = history.history

# 1 satırda 2 grafik olacak şekilde çizim alanı oluşturuluyor
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# --- Kayıp (Loss) grafiği ---
ax[0].plot(history["loss"], "s-", label="Training Loss")       # Eğitim kaybını çiz
ax[0].plot(history["val_loss"], "o-", label="Validation Loss") # Doğrulama kaybını çiz
ax[0].set_xlabel("Epochs")              # X ekseni etiketi
ax[0].set_ylabel("Loss")                # Y ekseni etiketi
ax[0].set_title("Train and Validation Loss")  # Başlık
ax[0].legend()                          # Grafik açıklama kutusu

# --- Doğruluk (Accuracy) grafiği ---
ax[1].plot(history["accuracy"], "s-", label="Training Accuracy")       # Eğitim doğruluğunu çiz
ax[1].plot(history["val_accuracy"], "o-", label="Validation Accuracy") # Doğrulama doğruluğunu çiz
ax[1].set_xlabel("Epochs")                # X ekseni etiketi
ax[1].set_ylabel("Accuracy")              # Y ekseni etiketi
ax[1].set_title("Training and Validation Accuracy")  # Başlık
ax[1].legend()                            # Grafik açıklama kutusu

# --- Tüm grafiklerin gösterilmesi ---
plt.show()
```

# 🧠 Kullanıcı Yorumuna Göre Duygu Tahmini

```python
# Kullanıcıdan alınan metni modele uygun biçimde işleyip tahmin yapan fonksiyon
def predict_sentiment(model, text, word_index, maxlen):
    
    # Girilen metni küçük harfe çevirip kelimelere ayırıyoruz, her kelimeyi indeksliyoruz
    encoded_text = [word_index.get(word, 0) for word in text.lower().split()]
    
    # Modelin beklediği uzunlukta padding uygulanıyor
    padded_text = pad_sequences([encoded_text], maxlen=maxlen)
    
    # Model ile tahmin yapılıyor
    prediction = model.predict(padded_text)
    
    # Skoru (0 ile 1 arası) döndürüyoruz
    return prediction[0][0]

# IMDb veri kümesinin kelime indeksini tekrar alıyoruz
word_index = imdb.get_word_index()

# Kullanıcıdan bir film yorumu alıyoruz
user_input = input("Film yorumu giriniz: ")

# Girilen yorumun duygu skorunu tahmin ediyoruz
sentiment_score = predict_sentiment(model, user_input, word_index, maxlen)

# Tahmin edilen skor 0.5’ten büyükse "olumlu", değilse "olumsuz"
if sentiment_score > 0.5:
    print(f"Tahmin sonucu olumlu -> skor: {sentiment_score:.2f}")
else:
    print(f"Tahmin sonucu olumsuz -> skor: {sentiment_score:.2f}")
```
