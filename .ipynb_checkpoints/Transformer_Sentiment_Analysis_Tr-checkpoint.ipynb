# ‚úÖ Gerekli K√ºt√ºphanelerin Y√ºklenmesi

```python
# Sayƒ±sal i≈ülemler i√ßin NumPy k√ºt√ºphanesi i√ße aktarƒ±lƒ±yor.
import numpy as np

# TensorFlow ve Keras mod√ºlleri i√ße aktarƒ±lƒ±yor.
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# G√∂rselle≈ütirme i√ßin Matplotlib kullanƒ±lacak.
import matplotlib.pyplot as plt

# Gereksiz uyarƒ± mesajlarƒ±nƒ± bastƒ±rmak i√ßin warnings mod√ºl√º kullanƒ±lƒ±yor.
import warnings
warnings.filterwarnings("ignore")
```

# ‚úÖ IMDb Verisinin Y√ºklenmesi ve ƒ∞≈ülenmesi

```python
# En sƒ±k kullanƒ±lan 10.000 kelime ile sƒ±nƒ±rlama yapƒ±lƒ±r.
max_features = 10000

# Her yorum en fazla 100 kelime uzunluƒüunda olacak ≈üekilde ayarlanƒ±r.
maxlen = 100

# IMDb veri seti y√ºklenir. Sadece en sƒ±k ge√ßen 10.000 kelime dahil edilir.
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Eƒüitim ve test verileri aynƒ± uzunlukta olacak ≈üekilde pad edilir (kƒ±sa olanlar <PAD> ile doldurulur).
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# IMDb verisinde kullanƒ±lan kelimelerin indeks kar≈üƒ±lƒ±klarƒ± alƒ±nƒ±r.
word_index = imdb.get_word_index()

# ƒ∞ndeks-kelime s√∂zl√ºƒü√º ters √ßevrilir. Keras‚Äôƒ±n √∂zel token‚Äôlarƒ± nedeniyle +3 yapƒ±lƒ±r.
reverse_word_index = {index + 3: word for word, index in word_index.items()}

# √ñzel semboller atanƒ±r:
reverse_word_index[0] = "<PAD>"
reverse_word_index[1] = "<START>"
reverse_word_index[2] = "<UNK>"
reverse_word_index[3] = "<UNUSED>"
```

# ‚úÖ Yorumlarƒ±n Decode Edilmesi ve G√∂sterilmesi

```python
# Sayƒ±sal verileri tekrar metin haline √ßevirir.
def decode_review(encoded_review):
    return " ".join([reverse_word_index.get(i, "?") for i in encoded_review])

# Eƒüitim setinden rastgele 3 yorum se√ßilir.
random_indices = np.random.choice(len(x_train), size=3, replace=False)

# Se√ßilen yorumlar ekrana yazdƒ±rƒ±lƒ±r.
for i in random_indices:
    print(f"Yorum: {decode_review(x_train[i])}")
    print(f"Etiket: {y_train[i]}")
    print()
```

# üß± TransformerBlock Sƒ±nƒ±fƒ±

```python
# Transformer bloƒüunun tanƒ±mƒ±
class TransformerBlock(layers.Layer):

    def __init__(self, embed_size, heads, dropout_rate=0.3):
        super(TransformerBlock, self).__init__()

        # √áok ba≈ülƒ± dikkat mekanizmasƒ± (self-attention)
        self.attention = layers.MultiHeadAttention(num_heads=heads, key_dim=embed_size)

        # Normalizasyon katmanlarƒ±
        self.norm1 = layers.LayerNormalization(epsilon=1e-6)
        self.norm2 = layers.LayerNormalization(epsilon=1e-6)

        # Feed forward aƒüƒ± (2 tam baƒülantƒ±lƒ± katman)
        self.feed_forward = models.Sequential([
            layers.Dense(embed_size * 4, activation="relu"),  # Gizli katman
            layers.Dense(embed_size)  # Giri≈ü boyutuna geri d√∂n√º≈ü
        ])

        # Overfitting'i azaltmak i√ßin dropout
        self.dropout1 = layers.Dropout(dropout_rate)
        self.dropout2 = layers.Dropout(dropout_rate)

    def call(self, x, training):
        # Self-attention uygulanƒ±r
        attention = self.attention(x, x, training=training)

        # ƒ∞lk normalizasyon ve residual connection
        x = self.norm1(x + self.dropout1(attention, training=training))

        # Feed forward + dropout + ikinci normalizasyon
        feed_forward = self.feed_forward(x)
        return self.norm2(x + self.dropout2(feed_forward, training=training))
```

# üì¶ TransformerModel

```python
# Kendi Transformer tabanlƒ± sƒ±nƒ±flandƒ±rma modelimizi tanƒ±mlƒ±yoruz
class TransformerModel(models.Model):

    def __init__(self, num_layers, embed_size, heads, input_dim, output_dim, dropout_rate = 0.2):
        super(TransformerModel, self).__init__()

        # üî§ Embedding katmanƒ±: kelimeleri sayƒ±lardan vekt√∂rlere d√∂n√º≈üt√ºr√ºr
        self.embedding = layers.Embedding(input_dim=input_dim, output_dim=embed_size)

        # üß± Birden fazla Transformer bloƒüu olu≈üturuluyor (√∂rneƒüin 4 katman)
        self.transformer_blocks = [
            TransformerBlock(embed_size, heads, dropout_rate)
            for _ in range(num_layers)
        ]

        # üìè Zaman boyutunu ortalayƒ±p sabit vekt√∂r √ßƒ±karmak i√ßin global average pooling
        self.global_avg_pooling = layers.GlobalAveragePooling1D()

        # üíß Overfitting'i √∂nlemek i√ßin dropout uygulamasƒ±
        self.dropout = layers.Dropout(dropout_rate)

        # üîö Sonu√ß katmanƒ±: 1 √ßƒ±ktƒ± √ºretir (sigmoid ile 0-1 arasƒ± deƒüer ‚Äî binary classification i√ßin)
        self.fc = layers.Dense(output_dim, activation="sigmoid")

    def call(self, x, training):
        # üî§ Giri≈ü verisini embedding vekt√∂rlerine √ßevir
        x = self.embedding(x)

        # üîÅ Her transformer bloƒüunu sƒ±rayla uygula
        for transformer in self.transformer_blocks:
            x = transformer(x, training=training)

        # üìè Zamansal boyutu ortalama alarak sabit uzunlukta tek bir vekt√∂re indir
        x = self.global_avg_pooling(x)

        # üíß Eƒüitim sƒ±rasƒ±nda dropout uygula (overfitting'e kar≈üƒ±)
        x = self.dropout(x, training=training)

        # üîö Son tahmin katmanƒ± (sigmoid √ßƒ±ktƒ±sƒ±)
        return self.fc(x)
```

# ‚öôÔ∏è Transformer Modeli Olu≈üturma ve Derleme

```python
# üî¢ Transformer mimarisi i√ßin temel hiperparametreleri tanƒ±mlƒ±yoruz
num_layers = 4            # Ka√ß adet Transformer bloƒüu kullanƒ±lacak
embed_size = 64           # Her kelimenin embedding vekt√∂r boyutu
num_heads = 4             # Multi-head attention'da ka√ß 'ba≈ü' kullanƒ±lacak
input_dim = max_features  # Modelin g√∂receƒüi maksimum kelime sayƒ±sƒ± (s√∂zl√ºk boyutu)
output_dim = 1            # Binary sƒ±nƒ±flandƒ±rma i√ßin tek bir √ßƒ±kƒ±≈ü (0 veya 1)
dropout_rate = 0.3        # Dropout oranƒ± (%30)

# üß† Transformer modelini olu≈üturuyoruz
model = TransformerModel(num_layers, embed_size, num_heads, input_dim, output_dim, dropout_rate)

# üèóÔ∏è Modelin input ≈üekli belirtilerek yapƒ±sƒ± olu≈üturuluyor (embedding + attention i√ßin gerekli)
model.build(input_shape=(None, maxlen))  # (None, maxlen): batch_size belirsiz, sabit dizi uzunluƒüu

# üõ†Ô∏è Modeli derliyoruz:
# - Kayƒ±p fonksiyonu: binary_crossentropy (√ß√ºnk√º √ßƒ±ktƒ± 0 veya 1)
# - Optimizasyon: Adam
# - Performans metriƒüi: Accuracy (doƒüruluk)
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

# üèãÔ∏è‚Äç‚ôÇÔ∏è Transformer Modelini Eƒüitme

```python
# üèÅ Modeli eƒüitiyoruz:
# - x_train: eƒüitim verileri
# - y_train: eƒüitim etiketleri
# - epochs=5: model veriyi 5 kez g√∂recek
# - batch_size=256: her seferinde 256 √∂rnekle aƒüƒ±rlƒ±klar g√ºncellenecek
# - validation_data: eƒüitim sƒ±rasƒ±nda doƒürulama (test) verileriyle kontrol edilecek
history = model.fit(
    x_train, y_train,              # Eƒüitim verisi ve etiketleri
    epochs=5,                     # 5 defa t√ºm veri √ºzerinden ge√ß
    batch_size=256,                # 256 veri ile bir batch olu≈ütur
    validation_data=(x_test, y_test)  # Doƒürulama i√ßin test verisi kullan
)
```

# üìà Eƒüitim S√ºrecinin G√∂rselle≈ütirilmesi 

```python
# history nesnesinin i√ßindeki ger√ßek verilere eri≈üiyoruz (sadece dictionary kƒ±smƒ± lazƒ±m)
history = history.history

# 1 satƒ±rda 2 grafik olacak ≈üekilde √ßizim alanƒ± olu≈üturuluyor
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# --- Kayƒ±p (Loss) grafiƒüi ---
ax[0].plot(history["loss"], "s-", label="Training Loss")       # Eƒüitim kaybƒ±nƒ± √ßiz
ax[0].plot(history["val_loss"], "o-", label="Validation Loss") # Doƒürulama kaybƒ±nƒ± √ßiz
ax[0].set_xlabel("Epochs")              # X ekseni etiketi
ax[0].set_ylabel("Loss")                # Y ekseni etiketi
ax[0].set_title("Train and Validation Loss")  # Ba≈ülƒ±k
ax[0].legend()                          # Grafik a√ßƒ±klama kutusu

# --- Doƒüruluk (Accuracy) grafiƒüi ---
ax[1].plot(history["accuracy"], "s-", label="Training Accuracy")       # Eƒüitim doƒüruluƒüunu √ßiz
ax[1].plot(history["val_accuracy"], "o-", label="Validation Accuracy") # Doƒürulama doƒüruluƒüunu √ßiz
ax[1].set_xlabel("Epochs")                # X ekseni etiketi
ax[1].set_ylabel("Accuracy")              # Y ekseni etiketi
ax[1].set_title("Training and Validation Accuracy")  # Ba≈ülƒ±k
ax[1].legend()                            # Grafik a√ßƒ±klama kutusu

# --- T√ºm grafiklerin g√∂sterilmesi ---
plt.show()
```

# üß† Kullanƒ±cƒ± Yorumuna G√∂re Duygu Tahmini

```python
# Kullanƒ±cƒ±dan alƒ±nan metni modele uygun bi√ßimde i≈üleyip tahmin yapan fonksiyon
def predict_sentiment(model, text, word_index, maxlen):
    
    # Girilen metni k√º√ß√ºk harfe √ßevirip kelimelere ayƒ±rƒ±yoruz, her kelimeyi indeksliyoruz
    encoded_text = [word_index.get(word, 0) for word in text.lower().split()]
    
    # Modelin beklediƒüi uzunlukta padding uygulanƒ±yor
    padded_text = pad_sequences([encoded_text], maxlen=maxlen)
    
    # Model ile tahmin yapƒ±lƒ±yor
    prediction = model.predict(padded_text)
    
    # Skoru (0 ile 1 arasƒ±) d√∂nd√ºr√ºyoruz
    return prediction[0][0]

# IMDb veri k√ºmesinin kelime indeksini tekrar alƒ±yoruz
word_index = imdb.get_word_index()

# Kullanƒ±cƒ±dan bir film yorumu alƒ±yoruz
user_input = input("Film yorumu giriniz: ")

# Girilen yorumun duygu skorunu tahmin ediyoruz
sentiment_score = predict_sentiment(model, user_input, word_index, maxlen)

# Tahmin edilen skor 0.5‚Äôten b√ºy√ºkse "olumlu", deƒüilse "olumsuz"
if sentiment_score > 0.5:
    print(f"Tahmin sonucu olumlu -> skor: {sentiment_score:.2f}")
else:
    print(f"Tahmin sonucu olumsuz -> skor: {sentiment_score:.2f}")
```
